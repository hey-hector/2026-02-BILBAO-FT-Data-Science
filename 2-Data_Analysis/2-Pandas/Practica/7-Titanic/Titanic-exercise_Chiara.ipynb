24. Using [`.pivot_table()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) to count how many women or men survived by class, call it `counted`.

Don't panic and read the documentation!




```python
# counted = titanic.pivot_table(values= "survived", index= "gender", columns= "pclass", aggfunc="count")
# ASÍ SACO EL NÚMERO TOTAL DE NOT NAN; LOS SUPERVIVIENTES (1) SE SACAN HACIENDO LA SUMA DE SURVIVED
counted = titanic.pivot_table(values= "survived", index= "gender", columns= "pclass", aggfunc="sum")
counted
```

25. Add a new column with the sum of survived men and women, call it `counted['total']` 


```python
counted = titanic.pivot_table(values= "survived", index= "gender", columns= "pclass", aggfunc="sum", margins=True, margins_name="counted['total']")
counted
```
