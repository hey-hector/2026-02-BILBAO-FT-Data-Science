## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
# MAL, DE PRUEBA
numero1 = float(input("Introduce un número:").replace("," , "."))
numero2 = float(input("Introduce otro número:").replace("," , "."))
operacion = input(("¿Qué operación quieres realizar?").lower())
operacion = operacion.replace("ó" , "o")

while operacion != "salir":

    if operacion == "multiplicacion":
        resultado = numero1 * numero2

    elif operacion == "division":
        if numero2 == 0:
            print("No se puede dividir un número por 0.")
            continue
        resultado = numero1 / numero2
    
    elif operacion == "suma":
        resultado = numero1 + numero2

    elif operacion == "resta":
        resultado = numero1 - numero2

    elif operacion == "potencia":
        resultado = numero1 ** numero2

    else:
        print("La operación introducida no se puede realizar.")
        break

if operacion == "salir":
        break
```

```python
# BIEN, SIN TODOS LOS REQUERIMIENTOS
while (True): # Se ejecuta siempre hasta que haya un bucle
    operacion = input(("¿Qué operación quieres realizar?").lower())
    operacion = operacion.replace("ó" , "o")

    if operacion == "salir":
        break
    
    numero1 = float(input("Introduce un número:").replace("," , "."))
    numero2 = float(input("Introduce otro número:").replace("," , "."))

    match operacion:
    
        case "multiplicacion":
            resultado = numero1 * numero2

        case "division":
            if numero2 == 0:
                print("No se puede dividir un número por 0.")
                continue
            resultado = numero1 / numero2
    
        case "suma":
            resultado = numero1 + numero2

        case "resta":
            resultado = numero1 - numero2

        case "potencia":
            resultado = numero1 ** numero2

        case _:
            print("La operación introducida no se puede realizar.")
            break
    
    print(f"El resultado de la operación {operacion} es {resultado}.")

print("Fin.")
```

```python
operaciones_realizadas = []

while (True):
    operacion = input(("¿Qué operación quieres realizar?").lower()) # Puedes pedirle que introduzca un número por cada operación (ej. sumar = 1)
    operacion = operacion.replace("ó" , "o")

    if operacion == "salir":
        break
    
    numero1 = float(input("Introduce un número:").replace("," , "."))
    numero2 = float(input("Introduce otro número:").replace("," , "."))

    if "multiplicacion" or "multiplicar":
        resultado = numero1 * numero2

    if "division" or "dividir":
        if numero2 == 0:
            print("No se puede dividir un número por 0. Vuelve a intentar otra operación.")
            continue
        resultado = numero1 / numero2
    
    if "suma" or "sumar":
        resultado = numero1 + numero2

    if "resta" or "restar":
        resultado = numero1 - numero2

    if "potencia" or "elevar":
        resultado = numero1 ** numero2

    else:
        print("La operación introducida no se puede realizar.")
        break
    
    print(f"El resultado de la operación {operacion} es {resultado}.")
    
    operaciones_realizadas.append(resultado)
    imprimir = input("¿Quieres imprimir los resultados que has obtenido hasta el momento?").lower()
    imprimir = imprimir.replace("í", "i")
    if imprimir == "si":
        print(operaciones_realizadas)

print("Fin.")
```
