```python
#PASO 1: Crear un Tablero como una lista de listas de 10x10 que contenga caracteres espacio: " " 
#*de internet*
tablero = [] #se empieza con una lista vacía y le añadiremos filas (i) y columnas (j) de un rango de 10
for i in range(10):
    fila = [] #se crea la lista vacía de filas
    for j in range(10): #este bucle 'for' está dentro para que haya una columna por cada fila - para que estén asociadas (o eso creo)
        fila.append(" ") #para cada columna (y en consecuencia, cada fila) se le añade el espacio - ¿al final de la lista?
    tablero.append(fila) #a la lista vacía 'tablero', se le añade el bucle entero al final
    
print(tablero)
    
#Así tenemos la lista "tablero" de 10 filas y 1o columnas (10 listas con 10 elementos) con espacios en blanco
```

```python
#PASO 2: Insertar dos barcos sustituyendo en el Tablero los espacios en blanco por el caracter "B":
    #* Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
    #* Uno de 3 casillas en vertical: casillas [3][3], [4][3], [5][3]
    
#Internet me recomendaba hacer un bucle pero no es necesario. Empecé usando el comando "insert" pero al hacer tablero.insert([1][0], "B") me salía la "B" fuera de la lista.

# Al final lo hacemos tablero[coordenada_i][coordenada_j]:
tablero[1][0] = "B"
tablero[1][1] = "B"
tablero[1][2] = "B"
tablero[1][3] = "B"
tablero[3][3] = "B"
tablero[4][3] = "B"
tablero[5][3] = "B" 
#Esto significa que la coordenada_i es la lista en la que está el elemento y la coordenada_j será la posición dentro de esa lista
'''
Esto es lo que más entendía de internet pero no es necesario hacer el bucle:
for tablero_con_barcos in tablero: #"tablero_con_barcos" entiendo que es una variable de "tablero" con la que queremos cambiar los elementos de las posiciones
    for i in range(len(tablero_con_barcos)): #"i" se refiere a las posiciones que hay en todo el rango de la longitud de "tablero_con_barcos". ¿El range no sobra?
        if tablero_con_barcos[1][0] == " ":
            tablero_con_barcos[1][0] = "B" #debería funcionar porque todos los elementos de cada lista, son strings
            #habría que ir añadiendo todo "elif"s con cada posición de los barcos
'''
print(tablero)
```

```python
'''PASO 3: Escribir una celda de código separada de las anteriores para simular que el usuario está efectuando un disparo: 
    *  Solicite al usuario dos coordenadas `i,j` mediante dos inputs
    *  Compruebe si esas coordenadas están ocupada por una casilla barco (que haya una en "B")
        * Si hay una "B":
            * Cambiar el contenido de esa posición por una "x"
            * Imprimir por pantalla el texto "Tocado en posición `i,j`"
        * Si hay un " ":
            * Cambiar el contenido de esa posición por una "o"
            * Imprimir por pantalla el texto "Agua"
    * Imprimir por pantalla el Tablero
* Para comprobar que funciona bien la celda de disparar ejecutarla varias veces metiendo distintas coordenadas
```

```python
#PASO 3:

coordenada_i = int(input("Meta la fila deseada. Valores 0-9:")) #fila (Nº de lista dentro de la lista)
coordenada_j = int(input("Meta la columna deseada. Valores 0-9:")) #columna (Nº de elemento dentro de la lista)
print(coordenada_i, coordenada_j) 
print(tablero[coordenada_i][coordenada_j])
if tablero[coordenada_i][coordenada_j] == "B": 
    tablero[coordenada_i][coordenada_j] = "x"
    print("Tocado en", coordenada_i, coordenada_j)
elif tablero[coordenada_i][coordenada_j] == " ":
    tablero[coordenada_i][coordenada_j] = "o"
    print("Agua")
print(tablero)
```
